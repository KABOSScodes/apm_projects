# Basic plots and histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

# plt.style.use(['science'])
# plt.style.use(['science', 'ieee']) # Need to install latex first
plt.style.use(['science', 'notebook', 'grid'])
```

```python
# Simple plots
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o',color='purple', lw=0.7, ms=5, label='Data')
plt.plot(x2,y2,'-',color='green', lw=0.7, ms=5, label='Theory')
plt.xlabel('time [s]')
plt.ylabel('voltage [V]')
plt.ylim(top=1.6)
plt.legend(loc='upper right', fontsize=10, ncol=2)
# plt.tight_layout() # Slightly alters the look of the plot. This command is probably  more needed when doing subplots. However, if you want to manually set vertical and horizontal 
# distances for subplots, you should avoid using tight_layout as it seems to override the manual settings.
plt.show()
```

```python
# Histograms/density plots
res = np.random.randn(1000)*0.2+0.4
res2 = np.random.randn(1000)*0.2+0.4
```

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True) # density=True normalizes plot so area under curve is 1
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True, label='res', histtype='step')
plt.hist(res2, bins=30, density=True, label='res2', histtype='step')
plt.legend(loc='upper right', fontsize=10, ncol=2)
plt.show()
```

# Subplots & More advanced API

```python
# Subplots
fig, axes = plt.subplots(2, 3, figsize=(12,12))
ax = axes[0][0]
ax.plot(x,y,'o',color='purple', lw=0.7, ms=5, label='Data')
ax.plot(x2,y2,'-',color='green', lw=0.7, ms=5, label='Theory')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Voltage [V]')
ax.set_ylim(top=1.5)
ax.text(0.1,0.1,'Text', transform=ax.transAxes, fontsize=10, color='red') # Transform is used to set coordinates relative to axes
ax.legend(loc='upper right', fontsize=10, ncol=2)
ax.set_title('Plot 1', fontweight='bold')
ax.tick_params(axis='both', which='major', labelsize=7, color='red') 

s = f"abe\nPer: {2:.3f}"
ax = axes[0][1]
ax.hist(res, bins=30, density=True, label='res', histtype='step')
ax.hist(res2, bins=30, density=True, label='res2', histtype='step')
ax.legend(loc='upper right', fontsize=10, ncol=2)
ax.set_title('Plot 2', fontweight='bold')
ax.set_ylim(top=2.35)
ax.text(0.1,0.5, s, transform=ax.transAxes, fontsize=10, bbox=dict(facecolor='white', edgecolor='black')) # Transform is used to set coordinates relative to axes

fig.text(0.5, 0.04, '$\Delta E$ Latex-tekst', ha='center', va='center', fontsize=15)

plt.subplots_adjust(hspace=0.15)  # Adjust space between subplots in vertical direction

plt.show()
```

```python
_ = np.linspace(-1,1,10)
x, y = np.meshgrid(_, _)
z = x**2 + x*y
```

```python
# Create a new figure
plt.figure(figsize=(6,3))

# Create a contour plot
plt.contourf(x, y, z, levels=30, vmin=0.3, cmap='PuOr') 

cbar = plt.colorbar()
cbar.set_label('asdf', size=10)
cbar.ax.tick_params(labelsize=10)  # Change '10' to the desired font size

plt.xticks([]) # Remove xticks

# Set the size of the y-axis labels
plt.tick_params(axis='y', labelsize=10)  # Change '10' to the desired font size

# Display the plot
plt.show()
```

```python
cs = plt.contour(x, y, z, levels=20) 
plt.clabel(cs, inline=1, fontsize=10)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [5, 6, 2, 3, 13]
z_scatter = [2, 3, 3, 3, 5]

# Sample data for surface plot
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x_scatter, y_scatter, z_scatter, color='r')

# Surface plot
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none', alpha=0.5)

# Labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter and Surface Plot')

# Display the plot
plt.show()
```
